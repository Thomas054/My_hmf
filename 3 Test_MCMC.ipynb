```python
from my_hmf import *
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
cosmo_params = {
    "H0": 70,
    "Om0": 0.294,
    "Ob0": 0.022 / 0.7**2,
    "ns": 0.965,
    "As": 2e-9
}

Ncamb = 1000
N_z = 5
zmax = 1
```

# MCMC

```python
s = Study(N_z,zmax, ["Om0","As"], knownpars = cosmo_params, Ncamb=Ncamb)
s.create_artificial_data(cosmo_params)
```

```python
N = 100
thetai = np.array([0.1, 1e-9])
step = 0.1*thetai

L_pars, L_chi2 = s.calc_params(thetai, N, step)
```

```python
%matplotlib widget
plt.figure()
plt.plot(L_pars[:,0], L_pars[:,1], 'o-')
plt.plot(L_pars[0,0], L_pars[0,1], '*', color='red')
plt.xlabel("Om0")
plt.ylabel("As")
plt.show()
plt.figure()
plt.plot(L_pars[:,0], L_chi2, 'o-')
plt.plot(L_pars[0,0], L_chi2[0], '*', color='red')
plt.xlabel("Om0")
plt.ylabel("chi2")
plt.show()
plt.figure()
plt.plot(L_pars[:,1], L_chi2, 'o-')
plt.plot(L_pars[0,1], L_chi2[0], '*', color='red')
plt.xlabel("As")
plt.ylabel("chi2")
plt.show()
```

```python
timestamp = time.strftime("%Y.%m.%d-%Hh%M")


np.savetxt(f'data/{timestamp}_pars.csv', L_pars)
np.savetxt(f'data/{timestamp}_chi2.csv', L_chi2)
```
