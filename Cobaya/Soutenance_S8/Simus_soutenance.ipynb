```python
from getdist import plots
import getdist
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

path = "C:/Users/thoma/Docs/Parcours_recherche/Code/Cobaya_data/Soutenance_S8/"
```

# 1

```python
run1_lmax1100 = getdist.loadMCSamples(path+'1_biasfree_P0fixed_lmax1100',
                                   settings={'ignore_rows':0.3} # Pour ignorer les 30% premières lignes
)

run1 = getdist.loadMCSamples(path+'1_biasfree_P0fixed',
                                   settings={'ignore_rows':0.3} # Pour ignorer les 30% premières lignes
)
```

```python
samples = [run1_lmax1100, run1]  # Liste des échantillons auxquels on s'intéresse
g = plots.get_single_plotter(chain_dir=path, width_inch=9)  # width_inch = largeur de la figure

# Pour faire joli
g.settings.figure_legend_frame = False  # Cadre autour de la légende
g.settings.alpha_filled_add = 0.4   # ?
g.settings.title_limit_fontsize = 14    # Taille de la police du titre
g.settings.legend_fontsize = 13   # Taille de la police de la légende
g.settings.axes_labelsize = 16  # Taille de la police des labels des axes
# g.settings.xtick.labelsize = 14
# g.settings.ytick.labelsize = 14



# Plot
g.triangle_plot(
    samples,
    ["omegam", "sigma8", "omch2","ombh2","ns","logA","H0", "bias_sz"],   # Paramètres à afficher
    filled=[True],
    legend_labels=["lmax1100","l pas borné"],  # 1 label par sample
    legend_loc="upper right",   # Position de la légende
    contour_colors=["black", "blue"],  # Couleurs des contours
    title_limit=1,
    markers={"x2": 0},
)


g.export('1.pdf')
```
