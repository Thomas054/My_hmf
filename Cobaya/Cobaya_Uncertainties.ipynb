```python
from getdist import plots
import getdist
```

# En réduisant l'incertitude sur $b$ et $\alpha$ séparément

```python
path = "C:/Users/thoma/Docs/Parcours_recherche/Code/Cobaya_data/"
run_normal = getdist.loadMCSamples(path+'Old/config_bias2',
                                   settings={'ignore_rows':0.3} # Pour ignorer les 30% premières lignes
)
run_b1=getdist.loadMCSamples(path+'config_b_uncertainty_1',
                            #   settings={'ignore_rows':0.3} # Pour ignorer les 30% premières lignes
)
run_b2=getdist.loadMCSamples(path+'config_b_uncertainty_2',
                            #   settings={'ignore_rows':0.3} # Pour ignorer les 30% premières lignes
)
run_a1=getdist.loadMCSamples(path+'config_alpha_uncertainty_2',
                              settings={'ignore_rows':0.3} # Pour ignorer les 30% premières lignes
)
run_a2=getdist.loadMCSamples(path+'config_alpha_uncertainty_4',
                              settings={'ignore_rows':0.3} # Pour ignorer les 30% premières lignes
)
```

### Sur $b$

```python
samples = [run_normal, run_b1,run_b2]  # Liste des échantillons auxquels on s'intéresse
g = plots.get_single_plotter(chain_dir=path, width_inch=9)  # width_inch = largeur de la figure

# Pour faire joli
g.settings.figure_legend_frame = False  # Cadre autour de la légende
g.settings.alpha_filled_add = 0.4   # ?
g.settings.title_limit_fontsize = 14    # Taille de la police du titre
g.settings.legend_fontsize = 13   # Taille de la police de la légende
g.settings.axes_labelsize = 16  # Taille de la police des labels des axes
# g.settings.xtick.labelsize = 14
# g.settings.ytick.labelsize = 14



# Plot
g.triangle_plot(
    samples,
    ["omegam", "sigma8"],   # Paramètres à afficher
    filled=[True],
    legend_labels=["original", "b/2", "b/4"],  # 1 label par sample
    legend_loc="upper right",   # Position de la légende
    contour_colors=["black", "blue", "red"],  # Couleurs des contours
    title_limit=1,
    markers={"x2": 0},
)


# g.export('testrun_triangle.pdf')
```

### Sur $\alpha$

```python
samples = [run_normal, run_a1,run_a2]  # Liste des échantillons auxquels on s'intéresse
g = plots.get_single_plotter(chain_dir=path, width_inch=9)  # width_inch = largeur de la figure

# Pour faire joli
g.settings.figure_legend_frame = False  # Cadre autour de la légende
g.settings.alpha_filled_add = 0.4   # ?
g.settings.title_limit_fontsize = 14    # Taille de la police du titre
g.settings.legend_fontsize = 13   # Taille de la police de la légende
g.settings.axes_labelsize = 16  # Taille de la police des labels des axes
# g.settings.xtick.labelsize = 14
# g.settings.ytick.labelsize = 14



# Plot
g.triangle_plot(
    samples,
    ["omegam", "sigma8"],   # Paramètres à afficher
    filled=[True],
    legend_labels=["original", r"$\alpha/2$", r"$\alpha/4$"],  # 1 label par sample
    legend_loc="upper right",   # Position de la légende
    contour_colors=["black", "green", "purple"],  # Couleurs des contours
    title_limit=1,
    markers={"x2": 0},
)


# g.export('testrun_triangle.pdf')
```
